```python
books = books[['ISBN','Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-L']]
books.head()
```

```python
# Now load the second dataset

users = pd.read_csv('dataset/BX-Users.csv', sep=";", on_bad_lines='skip', encoding='latin-1', low_memory=False)
users
```

```python
# Lets remane some columns name
users.rename(
    columns={
        "User-ID":'user_id',
        'Location':'location',
        "Age":'age'
    },
    inplace=True
)

users
```

```python
# Now load the third dataset

ratings = pd.read_csv('dataset/BX-Book-Ratings.csv', sep=";", on_bad_lines='skip', encoding='latin-1', low_memory=False)
ratings
```

```python
# Lets remane some columns name
ratings.rename(
    columns={
        "User-ID":'user_id',
        'Book-Rating':'rating'
    },
    inplace=True
)

ratings
```

```python
# Lets store users who had at least rated more than 200 books
x = ratings['user_id'].value_counts() > 200
```

```python
x
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
ratings.head()
```

```python
# Now join ratings with books
ratings_with_books = ratings.merge(books, on='ISBN')
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
number_rating
```

```python
number_rating.rename(columns={'rating':'number_of_ratings'}, inplace=True)
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
final_rating
```

```python
# Lets take those books which got at least 50 rating of user

final_rating = final_rating[final_rating['number_of_ratings'] >= 50]
final_rating
```

```python
# lets drop the duplicates
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
# Lets create a pivot table
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values= 'rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
# Get the ratings of the book at index 237 from all users
book_pivot.iloc[100,:]
```

```python
# Find 6 most similar books to the book at index 100 using KNN
# Output: 'distance' = similarity scores, 'suggestion' = indices of similar books
distance, suggestion = model.kneighbors(book_pivot.iloc[100, :].values.reshape(1, -1), n_neighbors=6)
```

## Model Testing

```python
def recommend_book(book_name):
    # Find the index of the given book in the pivot table
    book_id = np.where(book_pivot.index == book_name)[0][0]

    # Get distances and indices of 6 nearest neighbors (including the book itself)
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    print(f"You searched for: '{book_name}'\n")
    print("The suggested similar books are:\n")

    for i in range(1, len(suggestion[0])):  # skip index 0 because it's the same book
        suggested_book = book_pivot.index[suggestion[0][i]]
        similarity_score = distance[0][i]
        print(f"{i}. {suggested_book} (similarity score: {similarity_score:.4f})")
```

```python
book_name = "Call of the Wild"
recommend_book(book_name)
```
